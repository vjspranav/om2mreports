```python
import pandas as pd
import os
```

```python
# scan all dirs for respective .jtl files
x= ['retrieval', 'retrieval_lc', 'retrieval_rr']
# There is one jtl file in each directory, save the path to the file in a list
jtl_files = []
for i in x:
    for root, dirs, files in os.walk(i):
        for file in files:
            if file.endswith(".jtl"):
                 jtl_files.append(os.path.join(root, file))
```

```python
df_ret = pd.read_csv(jtl_files[0])
df_ret_lc = pd.read_csv(jtl_files[1])
df_ret_rr = pd.read_csv(jtl_files[2])

# Filter out all data which has responseCode != 200 || 201
df_ret = df_ret[df_ret['responseCode'].isin([200, 201])]
df_ret_lc = df_ret_lc[df_ret_lc['responseCode'].isin([200, 201])]
df_ret_rr = df_ret_rr[df_ret_rr['responseCode'].isin([200, 201])]
```

```python
# Get only rows which have allThreads as a multiple of 50
df_ret = df_ret[df_ret['allThreads'] % 50 == 0]
df_ret_lc = df_ret_lc[df_ret_lc['allThreads'] % 50 == 0]
df_ret_rr = df_ret_rr[df_ret_rr['allThreads'] % 50 == 0]
```

```python
df_ret.head()
```

```python
# Group by allThreads and get the mean of Latency
df_ret_diff = df_ret.groupby('allThreads')['Latency'].mean().reset_index()
df_ret_lc_diff = df_ret_lc.groupby('allThreads')['Latency'].mean().reset_index()
df_ret_rr_diff = df_ret_rr.groupby('allThreads')['Latency'].mean().reset_index()
```

```python
# Plot the graph using plotly with 300 dpi
import plotly.express as px
import plotly.graph_objects as go

# plot all three graphs in one figure
fig = go.Figure()

fig.add_scatter(x=df_ret['allThreads'], y=df_ret['Latency'], mode='lines', name='Retrieval')
fig.add_scatter(x=df_ret_lc['allThreads'], y=df_ret_lc['Latency'], mode='lines', name='Retrieval with Least Connected')
fig.add_scatter(x=df_ret_rr['allThreads'], y=df_ret_rr['Latency'], mode='lines', name='Retrieval with Round Robin')
fig.update_layout(
    autosize=False,
    width=1000,
    height=500,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
    paper_bgcolor="LightSteelBlue",
)

# Add legend for first graph
fig.update_layout(
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.01
    )
)

fig.update_xaxes(title_text='Number of Threads' )
fig.update_yaxes(title_text='Latency (ms)')
fig.show()
```
